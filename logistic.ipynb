```python
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(x_train,y_train)
```

```python
###Iteration Limit:

max_iter sets a limit on the number of iterations the solver is allowed to run. 
If the solver has not converged to the optimal solution within this number of iterations,
it will stop and return the current solution, 
along with a convergence warning.

Increasing max_iter allows the algorithm more time to find the optimal set of parameters. 
If your model is not converging with the default number of iterations,
increasing max_iter can help, especially for complex models or large datasets.

Convergence criteria :
Logistic regression can be optimized using various algorithms, including gradient descent, Newton’s method, and others. Each of these 
algorithms iteratively adjusts the model parameters to minimize the cost function (e.g., binary cross-entropy or log loss).

Convergence Criteria:

Parameter Stability: Convergence means that the changes in model parameters (weights) between iterations are smaller than a predefined tolerance.
    This indicates that the algorithm has reached a stable point where further iterations are unlikely to produce significant changes.

    Cost Function: Another convergence criterion is that the value of the cost function (the measure of model error) 
    has stabilized and is no longer significantly decreasing with additional iterations.
    ###
```

```python
model.score(x_test,y_test)
```
